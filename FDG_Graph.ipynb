```python
import pandas as pd
import hvplot.pandas
```

```python
df = pd.read_csv('sample_praise_set.csv')
```

```python
df.columns
```

Isolate a simple dataset with just to from and how much.

```python
df
```

```python
df = df[['To','From', 'IH per Praise']]
```

Fill nan values with the mean.

```python
df['IH per Praise'] = df['IH per Praise'].fillna(df['IH per Praise'].mean())
```

```python
weighted_edges = df.groupby(['To','From']).sum().reset_index()
```

```python
df.groupby('To')['IH per Praise'].sum().sort_values(ascending=False).hvplot.hist(bins=60, title='Total IH Received')
```

```python
df.groupby('From')['IH per Praise'].sum().sort_values(ascending=False).hvplot.hist(bins=60, title='Total IH Given')
```

```python
df.pivot_table(index='From',columns='To',values='IH per Praise',aggfunc=sum)
```

Making the nodes.

```python
nodes = set(df['To']).union(set(df['From']))
nodes = [{"id": i, "group": 1} for i in nodes]
nodes
```

Making the edges.

```python
weighted_edges.columns = ['target', 'source', 'value']

edges = weighted_edges.to_dict(orient='records')
```

Generating the data as json.

```python
data = {
    'nodes': nodes,
    'links': edges,
}

import json
with open('fdg/data.json', 'w') as fp:
    json.dump(data, fp)
```

To visualize the network, go into ./fdg and run: `python3 -m http.server`
