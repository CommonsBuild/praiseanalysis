```python
import pandas as pd
import plotly.express as px
```

### Import tags from hand labeled dataset

```python
df_tag = pd.read_excel('data/Praise to Label.xlsx', sheet_name="praise-tags",engine='openpyxl')
df_tag.set_index("Unnamed: 0", inplace=True)
df_tag
```

### Import and concatenate all data from hand labeled dataset

```python
data = []
for i in range(1, 57):
    df = pd.read_excel('data/Praise to Label.xlsx', sheet_name=f'new-sample-{i}',engine='openpyxl')
    data.append(df)
```

```python
hand_df = pd.concat(data)
```

```python
hand_df.rename(columns={"Unnamed: 0":"index", "Category Code":"hand_labeled"}, inplace=True)
```

```python
hand_df
```

```python
# hand_df.set_index("index", inplace=True)
hand_df = hand_df.iloc[:,[1,5]]
```

```python
hand_df
```

```python
hand_df.dropna(subset=["hand_labeled"], inplace=True)
```

```python
hand_df.set_index("Original Index", inplace=True)
```

```python
praise_df = pd.read_csv("data/cleaned_tagged_praise.csv", index_col=0)
praise_df
```

```python
hand_df.sort_index(na_position='first')
```

```python
merged_df = pd.merge(praise_df, hand_df, how="left", left_index=True, right_index=True)
```

```python
merged_df.dropna(subset=["hand_labeled"], inplace=True)
merged_df
```

### Export concatenated and merged dataset

```python
merged_df.to_csv("outputs/merged_hand_labeled.csv")
```

```python
merged_df
```

```python
total_IH_hand = merged_df.groupby("hand_labeled")["IH per Praise"].sum()
```

```python
total_IH_hand = total_IH_hand.sort_values(ascending=True)
```

```python
# total_IH_hand
tag_merged_hand = pd.merge(total_IH_hand, df_tag, how="left", left_index=True, right_index=True)
# tag_merged_hand.fillna("empty", inplace=True)
```

```python
tag_merged_hand.reset_index(inplace=True)
```

```python
tag_merged_hand
```

### Deal with different tags with the same category

```python
tag_merged_hand.iloc[12,2] = 'Participation, Retweets TEC7'
tag_merged_hand.iloc[14,2] = 'Participation, Retweets TE7'
tag_merged_hand.iloc[15,2] = 'Personal Praise TE8'
```

```python
tag_merged_hand
```

### Normalizing `IH per Praise` to have percentage

```python
tag_merged_hand["IH per Praise %"] = tag_merged_hand['IH per Praise'].copy()
tag_merged_hand["IH per Praise %"].fillna(0, inplace=True)
tag_merged_hand["IH per Praise %"] = (tag_merged_hand["IH per Praise %"]*100) / tag_merged_hand["IH per Praise %"].sum()
```

```python
px.bar(tag_merged_hand,
       x='Category',
       y='IH per Praise %',
       hover_data=['hand_labeled'],
      height=1200,
      title='Total Impact Hours per Category')
```
