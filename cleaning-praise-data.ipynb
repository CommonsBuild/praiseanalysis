#### 1. dropping emojis from "Room" names


We use the solution suggested by **jeffhale** in answer to this [this Stackoverflow post](https://stackoverflow.com/questions/51217909/removing-all-emojis-from-text).

```python
#uncomment to install the emoji library
#!pip install emoji
```

```python
#function written by jeffhale 
# source: https://stackoverflow.com/questions/51217909/removing-all-emojis-from-text

import emoji
import re

def strip_emoji(text):
    new_text = re.sub(emoji.get_emoji_regexp(), "", str(text))
    return new_text


noqdf.loc[:,"Room-NoEmoji"] = noqdf.loc[:,"Room"].apply(strip_emoji)
noqdf.head(5)
```

#### 3. Rename "PRAISE" as "Telegram" 

```python
noqdf["Server"].replace({"PRAISE": "Telegram"}, inplace = True)
noqdf.tail(5)
```

#### 4. Combining room and server 

```python
noqdf["Source"] = noqdf["Server"] + ":" + noqdf["Room-NoEmoji"]
noqdf.head(5)
```

Some dates have "nan", and others just have the word "dup" instead of a date. We may want to get rid of these in the future, but fr now, we will keep them since they may contain important information. 

Now we write a function that turns all of the dates in a standard format. 

```python
def normalize_dates(date_str):
  """
  This function takes in a string that contains a date and converts it to the 
  format Year-Month-Day.
  
  :param date_str: The string containing the date. 
  :return: The date in the format Year-Month-Day.
  """
  d = parser.parse(date_str)
  return d.strftime("%Y-%m-%d")
```

### Issues with Names

 There is a **naming** issue in the data frame that we need to address.


The issue is that some users receive praise with variations on their names. 

We will work first on making sure **each user is represented by their IH/CSTK** name. 

To do this, we use the "DO NOT TOUCH Imported" sheet that contains name information. 



```python
names_df = pd.read_excel('data/TEC Praise Quantification.xlsx',sheet_name="DO NOT TOUCH Imported",engine='openpyxl', usecols="A:D")
names_df.head()
```

```python
names_df.dropna(how='all').head(10)
```

#### We create a dictionary that matches each non-null entry in the spreadsheet to its "IH & CSTK Handle"

```python
source_cols = names_df.columns
names_dict = {} #create a blank dictionary

for i in range(len(names_df)):
    for col in source_cols:
        name_to_consider = names_df.loc[i, col]
        canonical_name = names_df.loc[i,"IH & CSTK Handle"]
        if pd.isna(canonical_name):
            canonical_name = name_to_consider
        if not(pd.isna(name_to_consider)):
            names_dict[name_to_consider] = canonical_name
            
```

```python
names_dict
```

#### Did we catch them all? See if there's anything in the combned user sets of "To" and "From" that isn't a key in the names_dict dictionary. 

```python
combined_users = set([]).union(set(noqdf["From"]),set(noqdf["To"]))

names_uncaught = sorted([str(x) for x in combined_users if (not(x in names_dict.keys()) and not(pd.isna(x)))])

print("All told, there are {} names in our data set that lack canonical representations: \n".format(str(len(names_uncaught))))
print("We do not have canonical representations for the following names: \n")
for name in names_uncaught:
    print(str(name))

```

#### OK, so there is some more to do here: 

1. 106 missing names is a lot.
2. Some of the missing are fairly obvious fixes:
- inconsistent case (zeptimusq vs. zeptimusQ, Ygganderson vs ygganderson)
- usernames with punctuation that get dropped when praising
- dropping the Discord four-digit identifier (AmwFund vs. AmwFund#0979)

The ones that are typos of these types can probably be fixed by casting the names to a bare-bones representation and checking to see if there is a reasonable key, though this has potential pitfalls (if two unrelated usernames are similar enough). Still, it's worth a shot. 

We're going to write a function to "clean" a name. 

```python
def clean_name(name):
    new_name = str(name)
    new_name = new_name.lower()
    new_name = re.sub('#\d\d\d\d','',new_name)
    new_name = re.sub('[^A-Za-z0-9]+', '', new_name) #remove all non-alphanumeric characters
    return new_name
```

Below are some sanity checks that the function works as intended.

```python
clean_name("AmwFund#0979")
```

```python
clean_name("Caeser (PST)#0046")
```

Now we use this function to check if a "clean" version of an unknown name has any good representation (and if so, we borrow it). 

```python
cleaned_keys = []
original_keys = list(names_dict.keys())
for key in original_keys:
    clean_key = clean_name(key)
    cleaned_keys.append(clean_key)
    if not(clean_key == key):
        names_dict[clean_key] = names_dict[key]
    
new_pairs = {}

for name in names_uncaught:
    clean_version = clean_name(name) #clean the name 
    key_matches = [key for key in cleaned_keys if clean_version == key] #make a list of all keys with same cleaned name
    if len(key_matches) > 0:
        for match in key_matches:
            if match in names_dict.keys():
                right_name = names_dict[match]
                new_pairs[name] = right_name
                names_dict[name] = right_name
                break
                    
print("We were able to add the following names: \n")
for name in new_pairs.keys():
    print(str(name) + "\t ----> \t" + str(names_dict[name]))
```

We will go through and see how many praises each of these unknown users have received. 

```python
def total_appearances(name):
    from_appearances = sum(noqdf["From"] == name)
    to_appearances = sum(noqdf["To"] == name)
    total = from_appearances + to_appearances
    return total 
```
